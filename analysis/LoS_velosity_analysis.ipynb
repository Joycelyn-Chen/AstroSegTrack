```python
import yt
import os
import argparse

hdf5_root = "/home/joy0921/Desktop/Dataset/200_360/finer_time_200_360_original"
timestamp = 206
posx_pc = -333
posy_pc = -443
posz_pc = -91

filename = f"sn34_smd132_bx5_pe300_hdf5_plt_cnt_0{timestamp}"
ds = yt.load(os.path.join(hdf5_root, filename))

prj = yt.ProjectionPlot(ds, "z", fields=("gas", "velocity_los"), weight_field=("gas", "density"),)
prj.annotate_timestamp()
prj.annotate_scale()
prj.annotate_sphere([posx_pc, posy_pc], radius=(10, "pc"), coord_system="plot") #, text=f"  {posz_pc} pc")

prj.show()
# prj.save(os.path.join(args.output_root, str(timestamp), f'{time_Myr}_prj.png'))
```
